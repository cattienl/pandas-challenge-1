## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
#View client 66037 total spending

view_client = df.loc[df["client_id"] == 66037, ["qty"]]
view_client.sum()
```

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df["subtotal"] = df["qty"] * df["unit_price"]
new_table = df[["qty","unit_price","subtotal"]]
new_table.head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def shipping_rate(qty):
    if qty > 50:
        return qty * 7
    return qty * 10
df["total_weight"] = df["qty"] * df["unit_weight"]   
df['shipping_price'] = df['total_weight'].apply(shipping_rate)

shipping_table = df[["unit_price","qty","unit_weight","total_weight","shipping_price"]]
shipping_table.head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df["total"] = (df["subtotal"] + df["shipping_price"]) * 0.0925

table_withtax= df[["subtotal","shipping_price","total"]]
table_withtax.head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df["line_price"] = (df["unit_price"]*df["qty"])+df["shipping_price"]
df["line_cost"] = (df["unit_cost"]*df["qty"])+ df["shipping_price"]
df.head()

#remove tax and total column
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['profit'] = df["line_price"] - df["line_cost"] + df["shipping_price"] + df["shipping_price"]
df.head(3)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.


client_grouped = df.groupby('client_id').agg(
    total_units_purchased=pd.NamedAgg(column='qty', aggfunc='sum'),
    total_shipping_price=pd.NamedAgg(column='shipping_price', aggfunc='sum'),
    total_revenue=pd.NamedAgg(column='line_price', aggfunc='sum'),
    total_profit=pd.NamedAgg(column='profit', aggfunc='sum')
)

sorted_clients = client_grouped.sort_values(by='total_profit', ascending=False)

top_5_clients = sorted_clients.head(5)

top_5_clients
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
# df["total_revenue"] = 
#df["total_revenue"] / 1000000
#df["total_profit"] = 
#df["total_profit"] / 1000000
#top_5_clients['total_revenue'] = top_5_clients['total_revenue'].map('${:,.2f}M'.format)
#top_5_clients['total_profit'] = top_5_clients['total_profit'].map('${:,.2f}M'.format)

top_5_clients = top_5_clients.rename(columns={"client_id": "Client ID", "total_units_purchased": "Units","total_shipping_price":"Shipping","total_revenue": "Revenue", "total_profit":"Profit"})
top_5_clients
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
sorted_df = top_5_clients.sort_values(by='Profit', ascending=False)
sorted_df
```
